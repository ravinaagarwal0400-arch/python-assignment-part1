```python
# -----------------------------------------
# TASK 1 - Data Parsing & Profile Cleaning
# -----------------------------------------

# Given Raw data -
raw_students = [
    {"name": "  ayesha SHARMA  ", "roll": "101", "marks_str": "88, 72, 95, 60, 78"},
    {"name": "ROHIT verma", "roll": "102", "marks_str": "55, 68, 49, 72, 61"},
    {"name": "  Priya Nair  ", "roll": "103", "marks_str": "91, 85, 88, 94, 79"},
    {"name": "karan MEHTA", "roll": "104", "marks_str": "40, 55, 38, 62, 50"},
    {"name": " Sneha pillai ", "roll": "105", "marks_str": "75, 80, 70, 68, 85"},
]

cleaned_students = [] # create an empty list tp append the cleaned data

print("\nTASK 1 OUTPUT\n")

# 1 - 

for student in raw_students:
    # remove spaces and convert to title case
    clean_name = student["name"].strip().title()

    # convert roll number string into integer
    roll_no = int(student["roll"])

    # split marks string and convert into integer list
    marks = [int(mark) for mark in student["marks_str"].split(", ")]

    student_data = {
        "name": clean_name,
        "roll": roll_no,
        "marks": marks
    }

    cleaned_students.append(student_data)

    # 2. validating name
    valid_name = all(word.isalpha() for word in clean_name.split())

    # printed in formated profile card

    print("=" * 40)
    print(f"Student : {clean_name}")
    print(f"Roll No : {roll_no}")
    print(f"Marks   : {marks}")
    print("=" * 40)

    if valid_name:
        print("✓ Valid name\n")
    else:
        print("✗ Invalid name\n")

# 4. print uppercase and lowercase for roll 103
for student in cleaned_students:
    if student["roll"] == 103:
        print("Uppercase :", student["name"].upper())
        print("Lowercase :", student["name"].lower())

## Task 1 gives us the cleared data to use for real world programming
```

```python
# -----------------------------------------
# TASK 2 - Marks Analysis
# -----------------------------------------

print("\n\nTASK 2 OUTPUT\n")

student_name = "Ayesha Sharma"
subjects = ["Math", "Physics", "CS", "English", "Chemistry"]
marks = [88, 72, 95, 60, 78]

# 1 for loop for printing grade label for each sujects - 

print(f"Student Name : {student_name}\n")

for i in range(len(subjects)):
    score = marks[i]

    if 90 <= score <= 100:
        grade = "A+"
    elif 80 <= score <= 89:
        grade = "A"
    elif 70 <= score <= 79:
        grade = "B"
    elif 60 <= score <= 69:
        grade = "C"
    else:
        grade = "F"

    print(f"{subjects[i]} : {score} -> {grade}")

# 2 - Calculations - 

total = sum(marks)
average = round(total / len(marks), 2)

highest = max(marks)
lowest = min(marks)

highest_subject = subjects[marks.index(highest)]
lowest_subject = subjects[marks.index(lowest)]

print("\nTotal Marks :", total)
print("Average Marks :", average)
print("Highest :", highest_subject, "-", highest)
print("Lowest :", lowest_subject, "-", lowest)


# while loop for adding subjects
new_subject_count = 0

while True:
    subject = input("\nEnter subject name (type done to stop): ")

    if subject.lower() == "done":
        break

    marks_input = input("Enter marks (0-100): ")

    if not marks_input.isdigit():
        print("Invalid input! Please enter numeric marks.")
        continue

    mark = int(marks_input)

    if mark < 0 or mark > 100:
        print("Marks should be between 0 and 100.")
        continue

    subjects.append(subject)
    marks.append(mark)
    new_subject_count += 1

print("\nNew subjects added :", new_subject_count)

updated_average = round(sum(marks) / len(marks), 2)
print("Updated Average :", updated_average)

```

```python
# -----------------------------------------
# TASK 3 - Class Performance Summary
# -----------------------------------------

print("\n\nTASK 3 OUTPUT\n")

class_data = [
    ("Ayesha Sharma", [88, 72, 95, 60, 78]),
    ("Rohit Verma", [55, 68, 49, 72, 61]),
    ("Priya Nair", [91, 85, 88, 94, 79]),
    ("Karan Mehta", [40, 55, 38, 62, 50]),
    ("Sneha Pillai", [75, 80, 70, 68, 85]),
]

pass_count = 0
fail_count = 0
class_avg_sum = 0
topper_name = ""
topper_avg = 0

print("Name              | Average | Status")
print("-" * 40)

for name, marks_list in class_data:
    avg = round(sum(marks_list) / len(marks_list), 2)

    if avg >= 60:
        status = "Pass"
        pass_count += 1
    else:
        status = "Fail"
        fail_count += 1

    print(f"{name:<18}| {avg:^7} | {status}")

    class_avg_sum += avg

    if avg > topper_avg:
        topper_avg = avg
        topper_name = name

class_average = round(class_avg_sum / len(class_data), 2)

print("\nStudents Passed :", pass_count)
print("Students Failed :", fail_count)
print("Class Topper :", topper_name, "-", topper_avg)
print("Class Average :", class_average)
```
